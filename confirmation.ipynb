```python
%matplotlib inline

import geopandas as gpd
import matplotlib.pyplot as plt

# change default figsize 
plt.rcParams['figure.figsize'] = (15, 12)
```

```python
# import data
bbox = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_50m_wgs84_bounding_box.geojson')
countries = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_110m_admin_0_countries.geojson')
```

```python
fig, ax = plt.subplots()

base_color = '#f0f0f0'
border_color = base_color

bbox.plot(ax=ax, color=base_color, zorder=0);
countries.plot(ax=ax, edgecolor=border_color, color='#ffffff', zorder=1);

ax.set(title="World map", 
       xlabel="X coordinates in degrees",
       ylabel="Y coordinates in degrees");
```

```python
%matplotlib inline

import geopandas as gpd
import matplotlib.pyplot as plt

# change default figsize 
plt.rcParams['figure.figsize'] = (15, 12)
```

```python
graticule = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_50m_graticules_15.geojson')
countries = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_50m_admin_0_countries.geojson')
pop_places = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_50m_populated_places.geojson')
```

```python
# verify CRSs (are all WGS84)
print(graticule.crs, countries.crs, pop_places.crs) # {'init': 'epsg:4326'} {'init': 'epsg:4326'} {'init': 'epsg:4326'}
```

```python
fig, ax = plt.subplots()

graticule.plot(ax=ax, color='#295896', linewidth=.4, zorder=1);
countries.plot(ax=ax, edgecolor='white', color='#cccccc', zorder=2);
pop_places.plot(ax=ax, color='#fd6a05', zorder=3, markersize=8);

ax.set(title="World map projected in WGS84", 
       xlabel="X coordinates in degrees",
       ylabel="Y coordinates in degrees");
```

```python
%matplotlib inline

import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15, 12)
```

```python
na_rails = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_10m_railroads_north_america.geojson')
us_states = gpd.read_file('http://www2.census.gov/geo/tiger/GENZ2017/shp/cb_2017_us_state_20m.zip')
```

```python
print(na_rails.crs) # output is {'init': 'epsg:4326'}
print(us_states.crs) # output is {'init': 'epsg:4269'}
```

```python
fig, ax = plt.subplots()

base_color = '#f0f0f0'
border_color = base_color
line_color = '#448ee4'

us_states.plot(ax=ax, edgecolor='white', color=base_color)
na_rails.plot(ax=ax, color=line_color, zorder=1);

ax.set(xlim=(-130,-65), ylim=(20,65));
```
